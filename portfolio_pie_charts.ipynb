# Portfolio Pie Charts
This notebook creates two pie charts:
- Main portfolio allocation (ETH, BTC, ATOM, RUNE, OP, Other)
- Breakdown of the "Other" assets


```python
import matplotlib.pyplot as plt

# Combine WETH amounts
weth_total = 19749.73 + 697664 + 4261 + 33000 + 20000

# Major holdings
major_assets = {
    "WETH": weth_total,
    "WBTC + cbBTC": 67736.69 + 609.43,
    "ATOM": 49000,
    "RUNE": 76000,
    "OP": 10730.65
}

# Minor holdings
minor_assets = {
    "GIV": 6938.48,
    "ARB": 4284.30,
    "ENS": 2261.66,
    "PEPE": 343.33,
    "EURC": 51460.96,
    "ETH": 19737.15,
    "rETH": 12731.80
}

# Calculate total portfolio value
total_value = sum(major_assets.values()) + sum(minor_assets.values())

# Plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 9))

# Major pie chart
ax1.pie(
    major_assets.values(),
    labels=[f"{k}\n${v:,.0f}" for k, v in major_assets.items()],
    autopct='%1.1f%%',
    startangle=140,
    textprops=dict(color="black", fontsize=10)
)
ax1.set_title('Major Portfolio Holdings', fontsize=14)

# Minor pie chart
ax2.pie(
    minor_assets.values(),
    labels=[f"{k}\n${v:,.0f}" for k, v in minor_assets.items()],
    autopct='%1.1f%%',
    startangle=140,
    textprops=dict(color="black", fontsize=10)
)
ax2.set_title('Minor Portfolio Holdings', fontsize=14)

# Display total below the charts
fig.text(0.5, 0.01, f'Total Portfolio Value: ${total_value:,.2f}', ha='center', fontsize=14)

plt.tight_layout(rect=[0, 0.03, 1, 1])
plt.show()
```
